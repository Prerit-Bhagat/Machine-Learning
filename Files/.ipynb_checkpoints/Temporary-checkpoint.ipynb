```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('data_stress_1.csv')
df.info()
columns_to_convert = ['snoring range', 'respiration rate', 'body temperature', 
                      'limb movement', 'blood oxygen ', 'eye movement', 
                      'hours of sleep', 'heart rate ']

# Convert columns to numeric, errors='coerce' will convert non-numeric values to NaN
df[columns_to_convert] = df[columns_to_convert].apply(pd.to_numeric, errors='coerce')
df.info()
```

```python
df.drop(['limb movement','eye movement'], axis=1,inplace=True)
```

```python
df.info()
```

```python
import seaborn as sns
corr_matrix = df.corr()
# Create a heatmap
plt.figure(figsize=(23, 15))
sns.heatmap(corr_matrix, annot=True)
plt.title('Correlation Heatmap')
plt.show()
```

```python
for i in df.columns:
    fig, axes = plt.subplots(1,1)
    sns.histplot(df[i], kde=True)
    # sns.distplot(df[i], kde = True, color ='red', bins = 30)
```

```python
from scipy.stats import skew

# Remove nan values
valid_data = df[~np.isnan(df)]

# Calculate skewness
skewness = skew(valid_data)

# Print skewness
print(f"Skewness of valid data: {skewness}")
```

```python
# Plotting the distplots without any transformation
from scipy import stats

for col in df.columns:
    plt.figure(figsize=(10,3))
    plt.subplot(1,2,1)
    sns.distplot(df[col])
    plt.title(col)

    plt.subplot(1,2,2)
    stats.probplot(df[col], dist="norm", plot=plt)
    plt.title(col)

    plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.compose import ColumnTransformer
from sklearn.feature_selection import SelectPercentile, chi2
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import PowerTransformer
from sklearn.model_selection import RandomizedSearchCV, train_test_split
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import KBinsDiscretizer
```

```python
df.columns=['snoringrange','respirationrate','bodytemperature','bloodoxygen','sleep','heartrate','stresslevels']
numeric_features = ["bodytemperature", "bloodoxygen","sleep","heartrate"]
bin_features=['snoringrange','respirationrate']
# mean_value = df['bodytemperature'].mean()  
# df['bodytemperature'].fillna(value=mean_value, inplace=True)
# mean_value = df['bloodoxygen'].mean() 
# df['bloodoxygen'].fillna(value=mean_value, inplace=True)
# mean_value = df['sleep'].mean() 
# df['sleep'].fillna(value=mean_value, inplace=True)
# mean_value = df['heartrate'].mean()
# df['heartrate'].fillna(value=mean_value, inplace=True)

# kbin_respirationrate = KBinsDiscretizer(n_bins=5,encode='ordinal',strategy='quantile')
bin_transformer = Pipeline(
    steps=[("imputer", SimpleImputer(strategy="mean"))]
)
# numeric_transformer = Pipeline(
#     steps=[("imputer", SimpleImputer(strategy="mean")), ("scaler", StandardScaler())]
# )

Power_transformer= Pipeline(
    steps=[ ("scaler", PowerTransformer(method='yeo-johnson',standardize=True)),("imputer", SimpleImputer(strategy="mean")),("binning", KBinsDiscretizer(n_bins=4,encode='ordinal',strategy='uniform')) ]
)
preprocessor2 = ColumnTransformer(
    transformers=[
        ("first",bin_transformer,slice(0,7)),
        ("num", Power_transformer,slice(0,7))
    ]
)

df.info()
```

```python
from sklearn import svm 
clf4=Pipeline(
    steps=[("preprocessor", preprocessor2),("classifier", svm.SVC(decision_function_shape='ovr'))]
)
clf4
```

```python
X=df.drop('stresslevels',axis=1)
y=df['stresslevels']
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.60, random_state=101)
X_train.info()
```

```python
clf4.fit(X_train,y_train)
```

```python
clf4.score(X_test,y_test)
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(clf4,X_train,y_train,cv=5,scoring='accuracy').mean()
```
