```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('data_stress_1.csv')
df.info()
```

```python
columns_to_convert = ['snoring range', 'respiration rate', 'body temperature', 
                      'limb movement', 'blood oxygen ', 'eye movement', 
                      'hours of sleep', 'heart rate ']

df[columns_to_convert] = df[columns_to_convert].apply(pd.to_numeric, errors='coerce')
df.drop(['limb movement','eye movement'], axis=1,inplace=True)
df.info()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.compose import ColumnTransformer
from sklearn.feature_selection import SelectPercentile, chi2
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import PowerTransformer
from sklearn.model_selection import RandomizedSearchCV, train_test_split
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import KBinsDiscretizer
```

```python
df.columns=['snoringrange','respirationrate','bodytemperature','bloodoxygen','sleep','heartrate','stresslevels']
numeric_features = ["bodytemperature", "bloodoxygen","sleep","heartrate"]
bin_features=['snoringrange','respirationrate']


# kbin_respirationrate = KBinsDiscretizer(n_bins=5,encode='ordinal',strategy='quantile')
impute_all=Pipeline(
    steps=[("imputer", SimpleImputer(strategy="mean")) ]
)
bin_transformer = Pipeline(
    steps=[("binning", KBinsDiscretizer(n_bins=5,encode='ordinal',strategy='uniform')) ]
)

Power_transformer= Pipeline(
    steps=[("scaler", PowerTransformer(method='yeo-johnson',standardize=True))]
)
preprocessor = ColumnTransformer(
    transformers=[
        
        ("first",bin_transformer,[0,1]),
        ("num", Power_transformer,slice(0,7))
    ]
)
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import BernoulliNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import MultinomialNB
clf = Pipeline(
    steps=[("impute",impute_all),("preprocessor", preprocessor), ("classifier",  MultinomialNB())]
)
clf2=Pipeline(
    steps=[("impute",impute_all),("preprocessor", preprocessor), ("classifier", BernoulliNB())]
)
clf3 = Pipeline(
    steps=[("impute",impute_all),("preprocessor", preprocessor), ("classifier", DecisionTreeClassifier(criterion='entropy'))]
)
clf4=Pipeline(
    steps=[("impute",impute_all),("preprocessor", preprocessor), ("classifier", RandomForestClassifier(criterion='entropy'))]
)
clf
```

```python
X=df.drop('stresslevels',axis=1)
y=df['stresslevels']
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.65, random_state=101)
X_train.info()
```

```python
clf4.fit(X_train,y_train)
```

```python
print("model score: %.3f" % (clf3.score(X_test, y_test)*100))
```
